### What does a Linear Regression  "look" like? 

```python
import numpy as np
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sns.set(style="darkgrid", palette="muted")
```

### Let's simulate some data to help us develop intuition.

```python
#this is our data generating process
def simulate_linear_data(N, beta_0, beta_1, eps_sigma_sq): 
    """
    Simulate a random dataset using a noisy
    linear process.
    
    Parameters
    ----------
    N: Number of data points to simulate
    beta_0: Intercept
    beta_1: Slope of univariate predictor, X
    
    Returns
    -------
        Pandas DataFrame
    """
    # Create a pandas DataFrame with column ’x’ containing # N uniformly sampled values between 0.0 and 1.0
    df = pd.DataFrame(
                {"x":
                    np.random.RandomState(42).choice(
                    list(map(
                        lambda x: float(x)/100.0, np.arange(100))), N, replace=False)
                } 
        )

    
    # Use a linear model (y ~ beta_0 + beta_1*x + epsilon) to
    
    eps_mean = 0.0
    epsilon = np.random.RandomState(42).normal(eps_mean, eps_sigma_sq, N)
    
    #Use a linear model (y ~ beta_0 + beta_1*x + epsilon)
    # generate a column ’y’ of responses based on ’x’
    df["y"] = beta_0 + beta_1*df["x"] + epsilon
    
    return df
```

### The frequentist approach: generate a 'line of best fit'

```python
beta_0 = 1.0 # Intercept
beta_1 = 2.0 # Slope

# Simulate 100 data points, with a variance of 0.5
N = 100
eps_sigma_sq = 0.5

# Simulate the "linear" data using the above parameters
df = simulate_linear_data(N, beta_0, beta_1, eps_sigma_sq)


#We use seaborn's linear model plot to show us our line
sns.lmplot(x="x", y="y", data=df, size=5, ci=None);
plt.xlim(0.0, 1.0);
```

## We said that for OLS standard error, Epsilon, is normally distributed. 

```python
#if that looks strange, try adjusting the number in RandomState to see how it affects the plot
sns.distplot(np.random.RandomState(42).normal(0., 0.5, 100));
```

## Bayesian Interpretation of Linear Regression: Monte Carlo approximations of linear Models

```python
import pymc3 as pm
basic_model = pm.Model()
#Comprehension Check: what does this mean?
#What is a context manager?
with basic_model:

    # Create the glm using the Patsy model syntax
    # We use a Normal distribution for the likelihood 
    pm.glm.GLM.from_formula("y ~ x", df, family=pm.glm.families.Normal())
    
    # Use Maximum A Posteriori (MAP) optimisation 
    # as initial value for MCMC
    start = pm.find_MAP()

    # Use the No-U-Turn Sampler
    # We use this step for continuous variables
    step = pm.NUTS()
    # Calculate the trace
    #Comprehension Check: What is happening here? 
    trace = pm.sample(5000, 
                      step, 
                      start, 
                      random_seed=42, 
                      progressbar=True)
```

# What is the Interpretation of the plots below?

```python
#What is the right column? What is the left column?
pm.traceplot(trace[500:])
plt.show()
```

## What is the meaning of the lines in this plot?

```python
#this line is plotting our points
sns.lmplot(x="x", y="y", data=df, size=10, fit_reg=False) 
plt.xlim(0.0, 1.0)
plt.ylim(0.0, 4.0)

#this is the series of thin lines
pm.plot_posterior_predictive_glm(trace, samples=100,
                              label='posterior predictive regression lines');

plt.title('Posterior predictive regression lines');

#gen x and y data to plot line
x = np.linspace(0, 1, N)
y = beta_0 + beta_1*x

#lets plot oour model's estimate
plt.plot(x, y, label="True Regression Line", lw=3., c="green") 
plt.legend(loc=0)
```
